```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Add Language and Id to kernel
lang = pd.read_csv('../Datasets/KernelLanguages.csv')
kern = pd.read_csv('../Datasets/KernelVersions.csv')
kernel = pd.read_csv('../Datasets/Kernels.csv')
kern = kern[['KernelId','KernelLanguageId']]
kern.drop_duplicates(subset='KernelId',keep='last',inplace=True)
lang = lang[['Id','DisplayName']]
lang.columns = ['KernelLanguageId','DisplayName']
out = pd.merge(kern,lang,on = 'KernelLanguageId')
out = out.sort_values('KernelId')
out.columns = ['Id','LanguageId','LanguageName']
df = kernel.merge(out,on = 'Id',how = 'inner')
df = df.drop(['CreationDate','EvaluationDate','MadePublicDate','MedalAwardDate','LanguageId'],axis = 1)
```

```python
# Fixing vote with 0 views
temp = df[['TotalViews','TotalVotes']].copy()
temp = temp[(temp.TotalVotes > 0) & (temp.TotalViews == 0)]
df.loc[temp.index,'TotalVotes'] = 0
```

```python
#Change Language type to numeric classes
df.LanguageName[df.LanguageName=="R"]=1
df.LanguageName[df.LanguageName=="Python"]=2
df
```

```python
#change the IsProjectLanguageTemplate to numeric datatype
df.IsProjectLanguageTemplate[df.IsProjectLanguageTemplate==False]=int(0)
df.IsProjectLanguageTemplate[df.IsProjectLanguageTemplate==True]=int(1)

df
```

```python
df['ForumTopicId'].fillna(0,inplace=True)
df
```

```python
#Export
df.to_csv('../Datasets/KernelsCleaned.csv',index=False)
kern = pd.read_csv('../Datasets/KernelsCleaned.csv')
kern.head()
```
